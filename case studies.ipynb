# 1. Read the data set into python environment.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
data=pd.read_csv('cars_data.csv')
data
```

# 2. To check null values present in the dataset.

```python
data.isna().any()
```

There is no null values present in the dataset

#  3. Bar graph of male vs female

```python
plt.figure(figsize=(9,7))
sns.countplot(data=data,x='Buyer Gender')
```

This graph shows that the number of female buyers is higher than the number of male buyers.

# 4. To find the top five cars based on their sales price

```python
dc=data.sort_values('Sale Price',ascending=False)
dc[['Make','Model','Sale Price']].head()
```

 Volkswagen Jetta III ,Audi A6,Lexus LX,Mazda MX-6 and Lexus ES are the top 5 cars

# 5. To find the least five cars based on their resell price

```python
dc=data.sort_values('Resell Price',ascending=True)
dc[['Make','Model','Resell Price']].head()
```

Chevrolet 1500,GMC Savana 1500,Jaguar XF,Mitsubishi 3000GT and infiniti G37 are least five cars based on their resell price
